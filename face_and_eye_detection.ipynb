```
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```
img=plt.imread("/content/IMG-20190301-WA0004new.jpg")
```

```
plt.imshow(img)
```

```
img.shape
```

```
#gray= cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
#colr=cv2.cvtColor(gray,cv2.COLOR_GRAY2BGR)
colr=cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
```

```
model=cv2.CascadeClassifier("/content/haarcascade_frontalface_alt.xml")
part=cv2.CascadeClassifier("/content/haarcascade_eye.xml")
```

```
faces=model.detectMultiScale(colr)
```

```
faces.shape
```

```
for face in faces:
  x,y,w,h=face
  facecrop=colr[y:y+h, x:x+w]
  fcol=img[y:y+h, x:x+w]
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),4)
  eyes=part.detectMultiScale(facecrop)
  for (ex,ey,ew,eh) in eyes:
    cv2.rectangle(fcol,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
```
